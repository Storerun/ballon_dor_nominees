```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ballon_dor = pd.read_csv('Ballon Dor - Sheet1.csv')
ballon_dor.head()
```

# Number  of nominees

```python
print('The number of nominees: ',ballon_dor.shape[0])
```

# Players by Club

```python
plt.figure(figsize=(12,8))

candidates_club = ballon_dor.Club.value_counts().sort_values()
candidates_club.plot(kind='barh')
plt.title('Players By Club')
plt.show()
```

# Players by position

```python
candidates_position = ballon_dor.Position.value_counts().sort_values()

plt.pie(candidates_position, labels =candidates_position.index,autopct = "%0.2f%%",shadow=True,textprops={"fontsize":15},radius=1.5,explode=[0.05,0.1,0.1,0.09] )

plt.show()
```

# Players By league

```python
ballon_dor.Club.unique()
```

```python
Epl=['Manchester City','Liverpool','Chelsea','Tottenham Hotspur']
LaLiga=['Real Madrid','Villarreal','FC Barcelona','Atletico Madrid']
Bundesliga=['Borussia Dortmund','Bayern Munich']
SeriaA= ['AC Milan', 'Inter Milan','Juventus']
Ligue1= ['PSG']

def league_identifier(team):
    if team in Epl:
        return 'EPL'
    elif team in LaLiga:
        return 'LaLiga'
    elif team in Bundesliga:
        return 'Bundesliga'
    elif team in SeriaA:
        return 'SeriaA'
    else:
        return 'Ligue1'
```

```python
ballon_dor['League'] = ballon_dor.Club.apply(lambda x:league_identifier(x) )
ballon_dor.head()
```
